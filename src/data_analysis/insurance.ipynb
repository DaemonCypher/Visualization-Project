```python
import pandas as pd

# Load the data
df = pd.read_csv("../../static/insurance.csv")
# Define functions for BMI category and charge tier
def bmi_category(bmi):
    if bmi < 18.5:
        return 1  # Underweight
    elif bmi < 25:
        return 2  # Normal weight
    elif bmi < 30:
        return 3  # Overweight
    else:
        return 4  # Obesity

def charge_tier(charge):
    if charge > 30000:
        return 3  # High
    elif charge > 15000:
        return 2  # Medium
    else:
        return 1  # Low
    
def gender(sex):
    return 1 if sex == "male" else 0 
def round(charges):
    return int(charges)
def region(region):
    if region == "southwest":
        return 1  
    elif region =="southeast":
        return 2  
    elif region =="northwest":
        return 3  
    else:
        return 4  
def smoke_category(smoker):
    return 1 if smoker == "yes" else 0  # 1 for smoker, 0 for non-smoker

# Apply the functions
df["tier"] = df["charges"].apply(charge_tier)
df["obesity"] = df["bmi"].apply(bmi_category)
df["bmi"] = df["bmi"].apply(round)
df["smoker"] = df["smoker"].apply(smoke_category)
df["sex"] = df["sex"].apply(gender)
df["charges"] = df["charges"].apply(round)
df["region"] = df["region"].apply(region)



# Save the updated dataframe
df.to_csv("matrix.csv", index=False)

print("New file saved as 'insurance_augmented.csv'")
```
